### Task 1

Create a new column in the DataFrame and name it `Total Sales`. This column is a sum of `Consulting Sales` and `Software Sales`. Use `head` or `tail` to peek at the resulting data frame to confirm that the output matches your expectation. What is the sum of the `Total Sales` column?

````python
clients = pd.read_csv("data_input/companies.csv", index_col=0)

clients[['Consulting Sales','Software Sales','Returns']]=\
clients[['Consulting Sales','Software Sales','Returns']].replace(
    '[^\d.]+', '', 
    regex=True)


#create sum value for each row
clients['Total Sales'] = clients["Consulting Sales"] + clients['Software Sales']
clients.head()
````
