```python
tc = pd.read_csv("data_input/techcrunch.csv")
```

```python
tc['round'].value_counts().head()
```

```python
pd.crosstab(tc['round'], columns = 'count').head()
```

```python
pd.crosstab(tc['round'], columns = tc.state, margins=True).head()
```

```python
pd.crosstab(tc['state'], tc['round'], normalize='index', margins=True)
```

```python
pd.crosstab(tc['state'], tc['round'], normalize='all', margins=True)
```

```python
cond1 = (tc['state'] == 'NY')
cond2 = (tc['state'] == 'CA')
nyca = tc.loc[(cond1) | (cond2), :]
pd.crosstab(index=nyca['category'], 
            columns=nyca['round'], 
            values=nyca['raisedAmt'], 
            aggfunc='mean')/1000000
```
