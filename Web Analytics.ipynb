```python
import pandas as pd
import numpy as np
```

```python
ana = pd.read_csv("data_input/analytics.csv", comment="#")
```

```python
ana = ana.head(11)
```

```python
ana = ana.fillna('-')
```

```python
ana.head(100)
```

```python
ana.dtypes
```

```python
ana.Users = ana.Users.str.replace(',', '').astype('int')
ana['New Users'] = ana['New Users'].str.replace(',', '').astype('int')
ana.Sessions = ana.Sessions.str.replace(',', '').astype('int')
ana['Goal Completions'] = ana['Goal Completions'].str.replace(',', '').astype('int')
ana['Bounce Rate'] = ana['Bounce Rate'].apply(lambda x: x.replace('%', '')).astype('float') / 100
ana['Goal Conversion Rate'] = ana['Goal Conversion Rate'].apply(lambda x: x.replace('%', '')).astype('float') / 100
ana['Goal Value'] = ana['Goal Value'].apply(lambda x: x.replace('$', '')).astype('float')
```

```python
ana.head()
```

```python
ana.dtypes
```

```python
ana['HiVol'] = ana['New Users']/ana['Users'] > 0.95
```

```python
ana.head(10)
```

```python
pd.crosstab(ana['Language'], columns = ana['HiVol'], values=ana['Sessions'], aggfunc='sum')
```

```python
cond1 = ana.Language.str.startswith('en')
ana.loc[cond1, 'Language'] = 'en'

cond1 = ana.Language.str.startswith('id')
ana.loc[cond1, 'Language'] = 'id'
```

```python
ana.head(15)
```

```python
pd.crosstab(ana['Language'], columns = ana['HiVol'], values=ana['Pages / Session'], aggfunc='mean').fillna(0).round(2)
```

```python
pd.crosstab(ana['Language'], columns = ana['HiVol'], values=ana['Sessions'], aggfunc='sum').fillna(0).round(2)
```

```python
pd.pivot_table(ana, values='Pages / Session', index = 'Language', columns = 'HiVol', aggfunc = 'mean')
```
